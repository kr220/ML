## Now we will apply GD to find the value of x at which fxn gives minimum value

### We need to update x taking some step size and gradient;
### When delta(X) almost becomes zero that woould be our ans

```python
X = 0  ## denoting by X that we need to find 
lr = 0.1 ## step size
error = []   # Contains values of Y at different wrong X(while updating X)

for i in range(50):
    grad = 2*(X-5) 
    temp = X
    X = X - lr*grad
    Y_at_this_X = (X-5)**2
    error.append(Y_at_this_X)
    if abs(temp - X) < 0.0001:
        break
    print(X)
```

```python
plt.plot(error)
```
