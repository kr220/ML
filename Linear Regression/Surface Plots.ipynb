```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6, 7])

a, b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a+b, cmap='coolwarm')  #(x,y,z)
plt.show()
```

```python
x = np.arange(-1, 1, 0.02)
y = x

x, y = np.meshgrid(x,y)
```

```python
fig2 = plt.figure()
ax = fig2.gca(projection='3d')
ax.plot_surface(x,y,x**2+y**2, cmap='rainbow')  #(x,y,z)
plt.show()
```

## Contour

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(x,y,x**2+y**2, cmap='rainbow')
plt.show()
```
